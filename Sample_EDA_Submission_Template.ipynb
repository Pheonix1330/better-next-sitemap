# **Project Summary -**

This project focuses on analyzing customer support data to understand customer satisfaction and identify the key factors that influence the Customer Satisfaction (CSAT) score. The dataset contains information related to customer interactions with support agents such as communication channel, product category, customer city, handling time, agent details, and CSAT score.

The main objective of this project is to perform Exploratory Data Analysis (EDA) to uncover patterns, trends, and relationships within the dataset and then build a Machine Learning model that can predict customer satisfaction scores. Understanding these patterns can help businesses improve their customer service quality and overall customer experience.

In the Exploratory Data Analysis phase, the dataset was first inspected to understand its structure, number of features, and types of variables. Data cleaning was performed by handling missing values, removing duplicate records, and correcting inconsistent data entries. Descriptive statistics were used to understand the distribution of numerical variables such as item price and connected handling time.

Several visualizations were created to gain insights from the data. Count plots were used to analyze the distribution of customer interactions across different communication channels such as chat, call, or email. Bar charts were used to identify the most common product categories that generate support requests. Box plots were used to understand how handling time affects customer satisfaction. Scatter plots were used to analyze relationships between numerical variables.

From the analysis, it was observed that longer handling times tend to reduce customer satisfaction. Certain product categories generate more complaints, indicating possible product or service issues. Agent-related factors such as tenure and shift also show some influence on CSAT scores. Additionally, some cities generate higher volumes of customer support interactions.

After completing the exploratory analysis, the dataset was prepared for machine learning by encoding categorical variables and selecting relevant features. A classification model was trained to predict the CSAT score based on the available features. Random Forest was used as the primary model because it handles both numerical and categorical data effectively and provides good performance for classification tasks.

The model was trained using a train-test split to evaluate its performance on unseen data. Accuracy was used as the primary evaluation metric. Feature importance analysis was also conducted to identify which variables most strongly influence customer satisfaction.

The results show that variables such as connected handling time, product category, communication channel, and agent experience significantly affect customer satisfaction levels. These insights can help companies improve their support processes by reducing response time, improving agent training, and addressing product-related issues.

Overall, this project demonstrates how data analysis and machine learning can be used to improve customer service quality and make data-driven business decisions.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Data Wrangling Code

# Remove duplicates
df = df.drop_duplicates()

# Fill missing values correctly
df["Customer Remarks"] = df["Customer Remarks"].fillna("No Remarks")

df["CSAT Score"] = df["CSAT Score"].fillna(df["CSAT Score"].median())

# Check remaining null values
df.isnull().sum()
````

### What all manipulations have you done and insights you found?

During the data wrangling process, several preprocessing steps were performed to make the dataset suitable for analysis. First, duplicate records were identified and removed to ensure data quality and avoid redundancy. Then, missing values in the dataset were handled appropriately. Missing values in the Customer Remarks column were replaced with a placeholder value such as "No Remarks", while missing values in the CSAT Score column were filled using the median value to maintain the overall distribution of the data.

Additionally, the dataset was inspected for incorrect data types and unnecessary columns, and the structure of the dataset was analyzed to ensure consistency. These steps helped in preparing the dataset for further exploratory data analysis and visualization.

From the cleaned dataset, several initial insights were observed. The dataset contains both categorical and numerical variables which provide valuable information about customer support interactions. Variables such as channel name, product category, agent shift, and handling time are likely to influence the CSAT Score. These features can help in identifying patterns related to customer satisfaction and service efficiency. Overall, the data wrangling process ensured that the dataset is clean, consistent, and ready for deeper analysis.
